```python
import mlflow.tensorflow
import numpy as np

model_uri = 'runs:/7b8d0740b0504040a7d63dbc787930aa/model'


# Load the model from MLflow
model = mlflow.tensorflow.load_model(model_uri)

# Print weights and biases for each layer
for i, layer in enumerate(model.layers):
    weights = layer.get_weights()
    if weights:  # Only layers with weights (e.g., Dense, Conv)
        print(f"Layer {i} ({layer.name}):")
        print(f"  Weights shape: {weights[0].shape}")
        print(f"  Weights (all):\n{weights[0]}")
        print(f"  Bias shape: {weights[1].shape}")
        print(f"  Bias (all):\n{weights[1]}")
        print("-" * 60)
    else:
        print(f"Layer {i} ({layer.name}): No weights/biases")
```

```python
for i, w in enumerate(weights):
    print(f"Layer {i}: shape={w.shape}, mean={w.mean()}, std={w.std()}")
```
